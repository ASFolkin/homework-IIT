##<font size="5" color="black" face="Times"><b>&emsp;5.4. Создание версии отчета в GIT</b></font><a name = 'part54'>

<font size="3" color="black" face="Times"><div align="justify">
&emsp;Добавляем в файл отчёта все изменения.<br>

##<font size="5" color="black" face="Times"><b>&emsp;5.5. Создание ветки отчета в GIT и добавление версии отчёта</b></font><a name = 'part55'>

<font size="3" color="black" face="Times"><div align="justify">
&emsp;Командой `git checkout -b second_try` создаём новую ветку и автоматическое переключаемся на неё.<br>
&emsp;Добавляем в файл отчёта все изменения.<br>
&emsp;Добавляем файл и сохраняем изменения с комментарием "Добавил шаг 5.4".<br>
</div></font>

##<font size="5" color="black" face="Times"><b>&emsp;5.6. Слияние изменений отчета в основную ветку</b></font><a name = 'part56'>

<font size="3" color="black" face="Times"><div align="justify">
&emsp;Командой `git checkout master` переключаемся на главную ветку.<br>
&emsp;Командой `git merge second_try` объединяем изменения с основной веткой.<br>
</div></font>

##<font size="5" color="black" face="Times"><b>&emsp;5.7. Перенос содержимого из локального репозитория на Github</b></font><a name = 'part57'>

<font size="3" color="black" face="Times"><div align="justify">
&emsp;Создаём репозиторий на <a href = 'https://github.com/ASFolkin/homework-IIT.git'>Github</a> с названием "homework-IIT".<br>
&emsp;Командой `git remote add origin <url>` добавляем новый удалённый репозиторий.<br>
&emsp;Командой `git push -u origin master` отправляем локальные изменения на удалённый репозиторий.<br>
</div></font>
